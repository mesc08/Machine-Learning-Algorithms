### KMEANS

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
pod = df.drop('Private',axis=1)
```

```python
kmeans.fit(pod)
```

```python
kmeans.cluster_centers_
```

### EVALUATION

```python
def coverter(private):
    if(private=='Yes'):
        return  1
    else:
        return 0
```

```python
df['Cluster']=df['Private'].apply(coverter)
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print("confusion matrix",confusion_matrix(df['Cluster'],kmeans.labels_))
```

```python
print("classification report", classification_report(df['Cluster'],kmeans.labels_))
```

```python
print("Accuracy Score ", accuracy_score(df['Cluster'],kmeans.labels_))
```
