```python
x = tf.constant(2)
y = tf.constant(3)
```

```python
with tf.Session() as sess:
    print("Operations done")
    print("Addition (x+ y) ",sess.run(x+y))
    print("Subtraction (x-y)", sess.run(abs(x-y)))
    print("Multiplication x*y ", sess.run(x*y))
    print("Division x/y ", sess.run(x/y))
```

```python
x = tf.placeholder(tf.int32)
y = tf.placeholder(tf.int32)
```

```python
x
```

```python
type(x)
```

```python
add = tf.add(x,y)
sub = tf.subtract(x,y)
mul = tf.multiply(x,y)
div = tf.div(x,y)
```

```python
d = {x: 30, y:30}
```

```python
with tf.Session() as sess:
    print("Operation with running")
    print("Adding ", sess.run(add, feed_dict=d))
    print("Subtracting ", sess.run(sub, feed_dict=d))
    print("Mulitply ", sess.run(mul, feed_dict=d))
    print("Division ", sess.run(div, feed_dict=d))
```

```python
import numpy as np
```

```python
a = np.array([[0.5, 0.5]])
b = np.array([[2.0],[2.0]])
```

```python
a
```

```python
a.shape
```

```python
mat1 = tf.constant(a, shape=[1,2])

mat2 = tf.constant(b, shape=[2,1])
```

```python
mmt = tf.matmul(mat1,mat2)
```

```python
with tf.Session() as sess:
    result = sess.run(mmt)
    print(result)
```
